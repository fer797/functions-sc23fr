```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('query_data.csv')

# Define a function to handle the specific datetime format
def parse_timestamp(ts):
    return pd.to_datetime(ts, format='%Y-%m-%dT%H:%M:%S.%fZ')

# Apply the function to the 'timestamp' column
data['timestamp'] = pd.to_datetime(data['timestamp'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Set the timestamp as the index
data.set_index('timestamp', inplace=True)

# Resample the data to 5-minute intervals, summing the values
resampled_data = data.resample('5T').sum()

# Plot the resampled data
plt.figure(figsize=(10, 6))
plt.plot(resampled_data.index, resampled_data['value'])
plt.title('5-Minute Interval Sum of Values')
plt.xlabel('Time')
plt.ylabel('Sum of Values')
plt.grid(True)
plt.show()
```
